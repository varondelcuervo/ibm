## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
import seaborn as sns
sns.boxplot(x="ConvertedComp", data=df)
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].describe()
```

```python
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
print(Q1, Q2)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR
lower_lim
```

```python
upper_lim
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
low = len(df['ConvertedComp']<lower_lim)
upp = len(df['ConvertedComp']>upper_lim)
low + upp
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
low = df['ConvertedComp']<lower_lim
upp = df['ConvertedComp']>upper_lim
#PARA ELIMINAR OUTLIERS
df1= df[~(low|upp)]
```

```python
sns.boxplot(x="ConvertedComp", data=df1)
```
